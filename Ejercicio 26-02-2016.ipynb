```python
#Bisección
import math
xl,xu,xr, fxl, fxr,i, xrant=0,1.3,0,1,1,1,0
es=0.5
def funcion(val):
    valor=math.pow(val,10)-1
    return valor
print ("Método Bisección")
print("{0:10} | {1:10} | {2:10} | {3:10} | {4:10} |".format("Iteracion", "xl", "xu","xr","Ea(%)"));
while ((fxl*fxr)!=0):
    xrant=xr
    xr=(xl+xu)/2
    fxl=funcion(xl)
    fxr=funcion(xr)
    ea=math.fabs((xr-xrant)/xr)*100
    print("{0:10} | {1:10} | {2:10} | {3:10} | {4:10} |".format(i,xl,xu,xr,round(ea,1)));
    if(ea<=es):
        break
    if((fxl*fxr)<0):
        xu=xr
    if((fxl*fxr)>0):
        xl=xr
    
    i+=1


#falsa posición
xl,xu,xr, fxl, fxr,i, xrant=0,1.3,0,1,1,1,0
print ("Falsa posición")
print("{0:10} | {1:10} | {2:10} | {3:10} | {4:10} |".format("Iteracion", "xl", "xu","xr","Ea(%)"));
while ((fxl*fxr)!=0):
    xrant=xr
    xr=xu-(funcion(xu)*(xl-xu))/(funcion(xl)-funcion(xu))
    ea=math.fabs((xr-xrant)/xr)*100
    print("{0:10} | {1:10} | {2:10} | {3:10} | {4:10} |".format(i,xl,xu,xr,round(ea,1)));
    
    if((fxl*fxr)<0):
        xu=xr
    if((fxl*fxr)>0):
        xl=xr
    i+=1
    if(ea<=es):
        break
```
