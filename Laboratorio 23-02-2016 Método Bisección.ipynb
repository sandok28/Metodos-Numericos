```python
    import math
    import matplotlib.pyplot as plt
    m,g,v,t,fxl,fx2,xl,xu,xr,i=68.1,9.8,40,10,1,1,12,16,0,1
    f,time=[],[]
    print("{0:10} | {1:10} | {2:10} | {3:10} | {4:10} | {5:10} |".format("Iteracion", "xl", "xu","Raiz", "F(xl)*F(xu)", "F(c)"));
    while((fxl*fx2)!=0):
        xr=(xl+xu)/2
        fxl=((g*m)/xl)*(1-math.exp(-(xl/68.1)*10.0))-v
        fx2=((g*m)/xr)*(1-math.exp(-(xr/68.1)*10.0))-v
        num= fxl*fx2
        fc=((g*m)/i)*(1-math.exp(-(i/68.1)*10.0))-v
        f.append(fc)
        time.append(i)
        print("{0:10} | {1:10} | {2:10} | {3:10} | {4:10} | {5:10} |".format(i,xl,xu,xr,num, fc));
        if((fxl*fx2)<0):
            xu=xr
        if((fxl*fx2)>0):
            xl=xr
        i+=1

    plt.plot(time,f)
    plt.xlabel('C')
    plt.ylabel('F(c)')
    plt.show()
```
