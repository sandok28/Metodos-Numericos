```python
import math
import matplotlib.pyplot as plt
m,g,v,t,fxl,fxr,xl,xu,xr,i=68.1,9.8,40,10,1,1,12,16,0,1
f,tiempo=[],[]
def funion(x):
    valor=((g*m)/x)*(1-math.exp(-(x/68.1)*10.0))-v
    return valor
print("{0:10} | {1:10} | {2:10} | {3:10} | {4:10} | {5:10} |".format("Iteracion", "xl", "xu","xr", "F(xl)*F(xu)", "Raiz", "F(c)"));
while((fxl*fxr)!=0):
    xrant=xr
    xr=xu-(funion(xu)*(xl-xu))/(funion(xl)-funion(xu))
    fxl=funion(xl)
    fxr=funion(xr)
    if((fxl*fxr)<0):
        xu=xr
    if((fxl*fxr)>0):
        xl=xr
    num= fxl*fxr
    fc=funion(i)
    f.append(fc)
    tiempo.append(i)
    print("{0:10} | {1:10} | {2:10} | {3:10} | {4:10} | {5:10} |".format(i,xl,xu,xr,num, fc));
    i+=1
plt.title()
plt.plot(tiempo,f)
plt.xlabel('C')
plt.ylabel('F(c)')
plt.show()
```
